```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

dfi = df.copy()

dfi['cut'] = label_encoder.fit_transform(dfi['cut'])
print(label_encoder.classes_)
dfi['color'] = label_encoder.fit_transform(dfi['color'])
print(label_encoder.classes_)
dfi['clarity'] = label_encoder.fit_transform(dfi['clarity'])
print(label_encoder.classes_)
```

```python
dfi.head()
```

```python
sns.pairplot(dfi)
```

```python
y_train = dfi[['price']].values
dfi.drop('price', axis=1, inplace=True)
X_train = dfi.values
```

```python
print(X_train[:5])
```

```python
print(y_train[:5])
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y = model.predict(X_train)
```

```python
print(y.shape)
print(y_train.shape)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train.reshape(-1))
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train))
```
