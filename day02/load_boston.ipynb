## load_boston

```python
import sklearn.datasets
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = sklearn.datasets.load_boston()
print(data.keys())
for f in data.keys():
    t = type(data[f])
    print('key: {}, type: {}'.format(f, t))
    if t == np.ndarray:
        print('shape: {}'.format(data[f].shape))
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df["CRIM"], hist=True, rug=True, label="CRIM")
plt.legend()
plt.show()
```

```python
sns.distplot(df["TAX"], hist=True, rug=True, label="TAX")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()
```

```python
sns.FacetGrid(df, hue='CHAS') \
.map(plt.scatter,'CRIM','TAX') \
.add_legend()
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='CHAS', y='CRIM',data=df)
```
