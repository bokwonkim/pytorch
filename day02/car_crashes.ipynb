```python
plt.figure(figsize=(15,8))
sns.barplot(x='abbrev', y='speeding', data=df)
```

```python
groupedvalues = df.groupby('abbrev').sum().reset_index()
groupedvalues.head()
```

```python
plt.figure(figsize=(15,8))
g = sns.barplot(x='abbrev',y='alcohol',data=groupedvalues)

for index, row in groupedvalues.iterrows():
    g.text(row.name, row.alcohol, round(row.alcohol, 2), color='black', ha="center")
```
