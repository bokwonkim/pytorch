## [Perceptron](https://scikit-learn.org/stable/modules/linear_model.html#perceptron)

```python
data = datasets.load_iris()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df.head()
```

```python
X_train = data.data
y_train = data.target
```

```python
from sklearn.linear_model import Perceptron

model = Perceptron(tol=1e-3, random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y = model.predict(X_train)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train))
```

## [Stochastic Gradient Descent](https://scikit-learn.org/stable/modules/linear_model.html#stochastic-gradient-descent-sgd)

```python
data = datasets.load_iris()
```

```python
X_train = data.data
y_train = data.target
```

```python
from sklearn.linear_model import SGDClassifier

model = SGDClassifier(max_iter=1000, tol=1e-3)
```

```python
model.fit(X_train, y_train)
```

```python
y = model.predict(X_train)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train))
```

## [Passive Aggressive Algorithms](https://scikit-learn.org/stable/modules/linear_model.html#passive-aggressive-algorithms)

```python
data = datasets.load_iris()
```

```python
X_train = data.data
y_train = data.target
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier

model = PassiveAggressiveClassifier(random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y = model.predict(X_train)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train))
```
