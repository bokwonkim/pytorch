## load_breast_cancer

```python
import sklearn.datasets
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = sklearn.datasets.load_breast_cancer()
print(data.keys())
for f in data.keys():
    t = type(data[f])
    print('key: {}, type: {}'.format(f, t))
    if t == np.ndarray:
        print('shape: {}'.format(data[f].shape))
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
sns.distplot(df["mean radius"], hist=True, rug=True, label="mean radius")
plt.legend()
plt.show()
```

```python
sns.distplot(df["mean texture"], hist=True, rug=True, label="mean texture")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()
```

```python
sns.FacetGrid(df, hue='target') \
.map(plt.scatter,'mean radius','mean texture') \
.add_legend()
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='target', y='mean radius',data=df)
```
